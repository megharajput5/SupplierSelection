```python
import csv
import random
import pandas as pd

# Generate a CSV file with 50000 records
def generate_supplier_data():
    product_types = ["Electronics", "Home Appliances", "Furniture", "Beauty Products"]
    supliers = ["Sup A","Sup B","Sup C"]
    communications = [1,2,3,4,5]
    with open("SupplierData.csv", "w", newline="") as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["SupplierName", "ProductType", "Quality", "Delivery",
                         "Price", "RepairService", "Reliability", "Communication", "Performance", "Reputation",
                         "Warranty", "Availability"])
        for i in range(50000):
            supplier_name = random.choice(supliers)
            product_type = random.choice(product_types)
            quality = random.choice([1,2,3,4,5])
            time = random.choice([2,4,6,8])
            price = random.randint(10, 1000)
            repair = random.choice([1,0])
            reliability = random.choice([1,2,3,4,5])
            communication = random.choice(communications)
            performance = random.choice([1,2,3,4,5])
            reputation = random.choice([1,2,3,4,5])
            warenty = random.choice([1,0])
            available = random.randint(10, 10000)

            writer.writerow([supplier_name,product_type,quality,time,price,repair,reliability,communication,
            performance,reputation,warenty,available])
    print("SupplierData.csv file has been generated.")
# Generate the CSV file
generate_supplier_data()
pd.read_csv("SupplierData.csv")
```
